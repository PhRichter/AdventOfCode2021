```python
import numpy as np

# Fixed input: Erste Zeile ist der Draw, danach n Spielbretter der Größe 5x5
def read_bingo(input_):
    with open(input_, "r") as f:
        cont = f.readlines()
        
    draw = np.array(cont[0].split(","), dtype=int)
    i=2
    boards = []
    while i < len(cont):
        boards.append(np.matrix(np.loadtxt(input_, skiprows=2, max_rows=5)))
        i+=6

    return draw, boards

def check_win(board):
    #Vertikal oder Horizontal
    i = 0
    while not (np.where(board[i,:] == -1)[-1].size == 5 \
              and np.where(board[:,i] == -1)[-1].size == 5 \
              and np.where(board[:,i] == -1)[-1].size == 5) or i < 5:
    for i in range(5):
        if np.where(board[i,:] == -1)[-1].size == 5:
            print(True)
        else:
            print(False)
        if np.where(board[:,i] == -1)[-1].size == 5:
            print(True)
        else:
            print(False)
        if np.where(board[:,i] == -1)[-1].size == 5:
            print(True)
        else:
            print(False)
    #Diagonal
    a = np.where(np.array([boards[0][i,i] for i in range(5)]) == -1)[0].size == 5
    print(a)
    #b = np.array([boards[0][4-i,i] for i in range(5)])
    
    
draw, boards = read_bingo("example.txt")

for number in draw:
    boards[0][np.where(boards[0] == number)] = -1
    check_win(boards[0])
    break
```
