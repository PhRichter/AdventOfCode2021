```python
import numpy as np

class DAY_19:
    def __init__(self, fname, verbose=True):
        self.__fname = fname
        self.__verbose = verbose
        self.__unique = []
        
    def __compare(self, arr_0, arr_1):
        return np.abs(arr_0[0]) in np.abs(arr_1) and np.abs(arr_0[1]) in np.abs(arr_1) and np.abs(arr_0[2]) in np.abs(arr_1)#: or \

    def __add(self, arr_0, arr_1, operator):
        out = [0,0,0]
        for i in range(3):
            if operator[i] == "+":
                out[i] = arr_0[i] + arr_1[i]
            elif operator[i] == "-":
                out[i] = arr_0[i] - arr_1[i]
        return out
        
    def read_scans(self):
        self.__beacons = []
        with open(self.__fname, "r") as f:
            cont = f.readlines()
            i = -1
            for line in cont:
                if "scanner" in line.split(",")[0]: 
                    self.__beacons.append([])
                    i+= 1
                #elif "scanner 2" in line.split(",")[0]: break
                try:
                    self.__beacons[i].append([int(line.split(",")[0]), int(line.split(",")[1]), int(line.split(",")[2])])
                except ValueError:
                    pass
                
        #if self.__verbose: print(self.__beacons[0])
        
    def add_coords(self, idx_0, idx_1, coords, operator):
        for i in range(len(self.__beacons[idx_0])):
            self.__beacons[idx_0][i] = self.__add(self.__beacons[idx_0][i], coords, operator)
        for i in range(len(self.__beacons[idx_1])):
            self.__beacons[idx_1][i] = self.__add(self.__beacons[idx_1][i], coords, operator)
            
    def calculate_differences(self, idx_0, idx_1):
        self.__idx_0 = idx_0
        self.__idx_1 = idx_1
        l = -1
        self.__diff = [[],[]]# for i in range(len(self.__beacons)]
        #Berechne die Differenzen
        for k in [idx_0, idx_1]:
            added = []
            l = l+1
            for i in range(len(self.__beacons[k])):
                for j in range(len(self.__beacons[k])):
                    if i == j or "{:d}{:d}".format(j+1, i+1) in added: continue #Verhinde doppeltes Berechnen der Differenzen. Überspringe, wenn die Beacons identisch sind
                    self.__diff[l].append([i+1,j+1, np.abs(np.array(self.__beacons[k][i]))-np.abs(np.array(self.__beacons[k][j]))])
                    added.append("{:d}{:d}".format(i+1,j+1))
        self.__same_0 = []
        self.__same_1 = []
        for i in range(len(self.__diff[0])):
            for j in range(len(self.__diff[1])):
                #Überprüfe dass die Differenz der Beacons übereinstimmt
                if self.__compare(self.__diff[0][i][2], self.__diff[1][j][2]):
                    #Speicher die Beacons ab, die in beiden Scans identische Abstände habe
                    self.__same_0.append(self.__diff[0][i][0])
                    self.__same_0.append(self.__diff[0][i][1])
                    self.__same_1.append(self.__diff[1][j][0])
                    self.__same_1.append(self.__diff[1][j][1])
            
        #Speicher die Zahlen ohne Duplikate
        self.__same_0_unique = list(set(self.__same_0))
        self.__same_1_unique = list(set(self.__same_1))
        print("Shared beacons", len(self.__same_0_unique))
        
    def __count(self, arr):
        pass
    
    def search_sonde_1(self):
        #Suche hier die Koordinaten von Sonde 1. Das müsste der Satz Koordinaten sein, der am häufigsten auftritt, (n mal bei n Beacons?)
        result_ppp = []
        result_mpp = []
        result_pmp = []
        result_ppm = []
        result_pmm = []
        result_mpm = []
        result_mmp = []
        result_mmm = []
        op = ""
        for i in range(len(self.__same_1)):
            x = self.__beacons[self.__idx_0][self.__same_0[i]-1]
            y = self.__beacons[self.__idx_1][self.__same_1[i]-1]
            #print(x,y)
            result_ppp.append(self.__add(x,y,"+++"))
            result_mpp.append(self.__add(x,y,"-++"))
            result_pmp.append(self.__add(x,y,"+-+"))
            result_ppm.append(self.__add(x,y,"++-"))
            result_pmm.append(self.__add(x,y,"+--"))
            result_mpm.append(self.__add(x,y,"-+-"))
            result_mmp.append(self.__add(x,y,"--+"))
            result_mmm.append(self.__add(x,y,"---"))
            result_pmp.append(self.__add(x,y,"+-+"))
        all_coord = []
        counted = {}
        operator = {}
        for i in result_ppp:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "+++"})
            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_mpp:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "-++"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_pmp:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "+-+"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_ppm:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "++-"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_pmm:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "+--"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_mpm:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "-+-"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_mmp:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i) : "--+"})

            else:
                counted[str(i)] += 1
            all_coord.append(i)
        for i in result_mmm:
            if str(i) not in counted.keys():
                counted.update({str(i) : 1})
                operator.update({str(i): "---"})
            else:
                counted[str(i)] += 1
            all_coord.append(i)
        #Finde die am häufigsten auftrenden Koordinaten. Das ist die Koordinate von Sonde_1 relative zu Sonde_0
        maximum = 0
        coord = ''
        #if self.__idx_1 == 4:
        #    for key in counted.keys():
        #        print(key, counted[key])
        for key in counted.keys():
            if counted[key] > maximum:
                maximum = counted[key]
                coord = key
        print(coord, operator[coord])

        coord_str = [int(coord.split("[")[1].split(",")[0]), int(coord.split("[")[1].split(",")[1]), int(coord.split("[")[1].split(",")[2].split("]")[0])] 
        for i in range(len(self.__beacons[self.__idx_1])):
            if i not in self.__same_1_unique:
                self.__unique.append(self.__add(coord_str, -np.array(self.__beacons[self.__idx_1][i-1]), operator[coord]))

                
    def print_num_beacons(self):
        print(len(self.__unique))
        
    def get_beacons(self):
        return self.__unique

day19_test = DAY_19("example.txt")
day19_test.read_scans()
#day19_test.add_coords(1, 4, np.array([68, -1246, -43]), "+-+")
day19_test.calculate_differences(0,1)
day19_test.search_sonde_1()
day19_test.print_num_beacons()
unique_beacons_1 = day19_test.get_beacons()
```
