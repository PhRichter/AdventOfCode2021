```python
par_open = ['(', '[', '{', '<']
par_close = [')', ']', '}', '>']
value = {'(': 3, ')': 3, '[': 57, ']': 57, '{': 1197, '}': 1197, '<': 25137, '>': 25137}
value_auto = {'(': 1, '[': 2, '{': 3, '<': 4}
class DAY_10:
    def __init__(self, log, verbose=True):
        self.__verbose = verbose
        self.__syntax_error = []
        self.__parentheses = []
        self.__score_auto = []
        with open(log, "r") as f:
            self.__parentheses = f.readlines()
    
        if self.__verbose: print(self.__parentheses)
        
    def iterate_parentheses(self):
        for parentheses in self.__parentheses:
            self.__stack = ""
            corrupted = False
            for par in parentheses.rstrip():
                #print(par, self.__stack)
                if par in par_open: 
                    self.__stack += par
                elif value[par] == value[self.__stack[-1]]:
                    self.__stack = self.__stack[:-1]
                else:
                    self.__syntax_error.append(value[par])
                    corrupted = True
                    #print("break ", par, self.__stack, value[par])
                    break
            if not corrupted:
                score_auto = 0
                if self.__verbose: print(self.__stack[::-1])
                for par in self.__stack[::-1]:
                    score_auto *= 5 
                    score_auto += value_auto[par]
                if self.__verbose: print(score_auto)
                self.__score_auto.append(score_auto)
    
    def show_syntax_error(self):
        print(self.__syntax_error, sum(self.__syntax_error))
        
    def show_score_autocomplete(self):
        self.__score_auto = sorted(self.__score_auto)
        print(self.__score_auto, self.__score_auto[int(len(self.__score_auto)/2)])
            
day10_test_1 = DAY_10("example.txt")
day10_test_1.iterate_parentheses()
day10_test_1.show_syntax_error()

day10_puzzle_1 = DAY_10("input.txt", False)
day10_puzzle_1.iterate_parentheses()
day10_puzzle_1.show_syntax_error()
```
