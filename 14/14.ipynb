```python
class DAY_14:
    def __init__(self, fname, verbose=True):
        self.__fname = fname
        self.__verbose = verbose
        
    def read_instructions(self):
        self.__to_insert = []
        self.__edukt = []
        self.__counter = {}
        self.__connections = {}
        with open(self.__fname, "r") as f:
            self.__template = f.readline().rstrip()
            content = f.readlines()
            
        for line in content[1:]:
            self.__to_insert.append(line.split("->")[1].rstrip().lstrip())
            self.__connections.update({line.split("->")[0].rstrip(): 0})
            self.__edukt.append(line.split("->")[0].rstrip())
        if self.__verbose: print(self.__to_insert)
        if self.__verbose: print(self.__edukt)
        
        #Counter fuellen
        #self.__counter = np.zeros(len(self.__edukt_set))
        #for letter in self.__template:
        #    if letter in self.__counter.keys():
        #        self.__counter[letter] += 1
        #    else:
        #        self.__counter.update({letter: 1})
                
        #Connections fuellen
        for i in range(len(self.__template)-1):
            self.__connections.update({self.__template[i:i+2]: 1})
            
        #Instructions fuellen
        self.__instructions = {}
        for i in range(len(self.__edukt)):
            self.__instructions.update({self.__edukt[i] : [self.__edukt[i][0]+self.__to_insert[i], self.__to_insert[i]+self.__edukt[i][1]]})
            
        #Nur Produkte
        self.__to_insert_set = sorted(list(set(self.__to_insert)))
        
        #Counter fuellen
        self.__counter = {}
        for letter in self.__template:
            if letter in self.__counter.keys():
                self.__counter[letter] += 1
            else:
                self.__counter.update({letter: 1})
            #self.__counter.update({[self.__edukt[self.__to_insert.index(letter)]] += 1
            
        print(self.__counter)
        print(self.__connections)
        print(self.__instructions)
        #print(self.__products)
        print(self.__to_insert_set)
        
        
    def create_polymer_puzzle_2(self, steps):
        for k in range(steps):
            new_connections = self.__connections.copy()
            to_update = []
            diff = []
            for connection in new_connections.keys():
                if new_connections[connection] != 0:
                    to_update.append(connection)
                    diff.append(self.__connections[connection])
                    letter_insert = self.__to_insert[self.__edukt.index(connection)]
                    if letter_insert in self.__counter.keys():
                        self.__counter[letter_insert] += new_connections[connection]
                    else:
                        self.__counter.update({letter_insert: 1})
                    #print(self.__counter[self.__edukt.index(connection)])
                    #print(self.__edukt.index(connection))
            for i in range(len(to_update)):
                new_connections[self.__instructions[to_update[i]][0]] += self.__connections[to_update[i]]
                new_connections[self.__instructions[to_update[i]][1]] += self.__connections[to_update[i]]
            for i in range(len(to_update)):
                new_connections[to_update[i]] -= diff[i]
            self.__connections = new_connections.copy()
        print(self.__connections)
        print(self.__counter)
        
    def create_polymer_puzzle_1(self, steps):
        self.__polymer = ""
        current_templ = self.__template
        for k in range(steps):
            polymer_steps = []
            for i in range(len(current_templ)-1):
                cur = current_templ[i:i+2].rstrip() 
                #Finde Anweisung passend zu den beiden Buchstaben
                for j in range(len(self.__edukt)):
                    if self.__edukt[j] == cur:                
                        polymer_steps.append(cur[0]+self.__to_insert[j]+cur[1])
            self.__polymer = self.__concat(polymer_steps)
            current_templ = self.__polymer
        if self.__verbose: print(self.__polymer)
        
    def __concat(self, steps):
        concated = ""
        for step in steps:
            if len(concated) > 0 and concated[-1] == step[0]:
                concated = concated[:-1]
            concated += step
        return concated
    
    def count_puzzle_1(self):
        if self.__verbose: print(list(set(self.__to_insert)))
        result = []
        for element in list(set(self.__to_insert)):
            res = 0
            for letter in sorted(self.__polymer):
                if element == letter: res += 1
            result.append(res)
        print(result, max(result)-min(result))
        
    def count_puzzle_2(self):
        print(max(sorted(self.__counter.values()))-min(sorted(self.__counter.values())))
            
day14_test_1 = DAY_14("example.txt")
day14_test_1.read_instructions()
day14_test_1.create_polymer_puzzle_1(10)
day14_test_1.count_puzzle_1()

day14_puzzle_1 = DAY_14("input.txt", False)
day14_puzzle_1.read_instructions()
day14_puzzle_1.create_polymer_puzzle_1(10)
day14_puzzle_1.count_puzzle_1()
```
