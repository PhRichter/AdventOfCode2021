```python
import numpy as np


class DAY_15:
    def __init__(self, fname, verbose=True, size=10):
        self.__fname = fname
        self.__verbose = verbose
        self.__size = size
        self.__q = []#Alle besuchbaren Punkte
        self.__q_val = []
        
    def read_input(self):
        self.__cavern_map = np.ones((self.__size, self.__size))
        cavern = np.loadtxt(self.__fname, dtype=str)
        for x in range(self.__size):
            for y in range(self.__size):
                self.__cavern_map[x,y] = int(cavern[x][y])

        if self.__verbose: print(self.__cavern_map)
        self.__cavern_map[0,0] = 0
        self.__q.append([1,0])
        self.__q.append([0,1])
        self.__q_val.append(self.__cavern_map[self.__q[0][0], self.__q[0][1]])
        self.__q_val.append(self.__cavern_map[self.__q[1][0], self.__q[1][1]])
        self.__not_visited = -1*np.ones((self.__size, self.__size))
        self.__not_visited[0,0] = 1
        
    def next_node(self):
        print(self.__q_val)
        for i in range(self.__size**2+1):
            #Suche den Punkt, welcher am günstigsten zu erreichen ist
            min_q_val = np.min(self.__q_val)
            index_of_min_q = self.__q_val.index(min_q_val)#Problem wenn der kleinste Index häufiger auftritt?
            index_of_next_node = self.__q[index_of_min_q]


            #Entferne den gefundenen Punkt aus der Q-Liste
            popped_q = self.__q.pop(index_of_min_q)
            popped_q_val = self.__q_val.pop(index_of_min_q)
                
            #Teste ob dieser Punkt das Ziel ist
            if index_of_next_node[0] == self.__size-1 and index_of_next_node[1] == self.__size-1: 
                print("finished", popped_q_val)
                break

            #Füge die neuen Elemente ein
            #if i > (self.__size-1)**2: print(i, min_q_val)
            if index_of_next_node[0]+1 < self.__size and self.__not_visited[index_of_next_node[0]+1,index_of_next_node[1]] == -1:
                self.__q.append([index_of_next_node[0]+1,index_of_next_node[1]])
                self.__q_val.append(min_q_val + self.__cavern_map[index_of_next_node[0]+1,index_of_next_node[1]])
                self.__not_visited[index_of_next_node[0]+1,index_of_next_node[1]] = 1
                
            if index_of_next_node[0]-1 >= 0 and self.__not_visited[index_of_next_node[0]-1,index_of_next_node[1]] == -1:
                self.__q.append([index_of_next_node[0]-1,index_of_next_node[1]])
                self.__q_val.append(min_q_val + self.__cavern_map[index_of_next_node[0]-1,index_of_next_node[1]])
                self.__not_visited[index_of_next_node[0]-1,index_of_next_node[1]] = 1

            if index_of_next_node[1]+1 < self.__size and self.__not_visited[index_of_next_node[0],index_of_next_node[1]+1] == -1:
                self.__q.append([index_of_next_node[0],index_of_next_node[1]+1])
                self.__q_val.append(min_q_val + self.__cavern_map[index_of_next_node[0],index_of_next_node[1]+1])
                self.__not_visited[index_of_next_node[0],index_of_next_node[1]+1] = 1

            if index_of_next_node[1]-1 >= 0 and self.__not_visited[index_of_next_node[0],index_of_next_node[1]-1] == -1:
                self.__q.append([index_of_next_node[0],index_of_next_node[1]-1])
                self.__q_val.append(min_q_val + self.__cavern_map[index_of_next_node[0],index_of_next_node[1]-1]) 
                self.__not_visited[index_of_next_node[0],index_of_next_node[1]-1] = 1

           
        print(index_of_next_node, i, min_q_val)
        
day15_test_1 = DAY_15("example.txt", True, 10)
day15_test_1.read_input()
day15_test_1.next_node()

import datetime as dt
start_time = dt.datetime.now()
day15_puzzle_1 = DAY_15("input.txt", size=100)
day15_puzzle_1.read_input()
day15_puzzle_1.next_node()
print((dt.datetime.now()-start_time).total_seconds())
```
