```python
def can_be_visited(visited_cave, counter):
    if visited_cave == "start":
        return False
    if visited_cave.lower() != visited_cave:
        return True
    #Teste ob es sich um eine kleine Höhle handelt
    if visited_cave.lower() == visited_cave:
        #Wurde diese Höhle schon besucht?
        if counter[visited_cave] == 0:
            return True
        #Überprüfe, ob schon eine kleine Höhle besucht wurde
        for cave in counter.keys():
            if cave.lower() == cave and counter[cave] > 1:
                return False
        return True

class DAY_12:
    def __init__(self, fname, verbose=True, puzzle=1):
        self.__fname = fname
        self.__verbose = verbose
        self.__puzzle = puzzle
        
    def read_map(self):
        self.__map = []
        with open(self.__fname, "r") as f:
            self.__map = f.readlines()
            
        if self.__verbose: print(self.__map)

    #Zähle die Höhlen und setze alle Besuchszähler auf 0
    def count_caves(self):
        caves = []
        self.__cave_counter = {}
        for cave in self.__map:
            caves.append(cave.split("-")[0].rstrip())
            caves.append(cave.split("-")[1].rstrip())
            
        if self.__verbose: print(list(set(caves)))
        
        for cave in list(set(caves)):
            self.__cave_counter.update({cave: 0})
        
        if self.__verbose: print(self.__cave_counter)
            
    def __next_cave(self, current, counter, path):
        counter['start'] = 1
        path = path + "," + current
        for cave in self.__map:
            left_cave = cave.split("-")[0].rstrip()
            right_cave = cave.split("-")[1].rstrip()
            #Wenn die aktuelle Höhle das Ziel ist, speichere den gefundenen Pfad ab
            if current == "end":
                self.__paths.append(path)
            #Gehe in die nächste Höhle, erhöhe den Zähler und reduziere ihn, damit die Höhle beim Prüfen 
            #eines anderen Weges besuchbar bleibt
            elif left_cave == current and \
            (self.__puzzle == 1 and not (right_cave.lower() == right_cave and counter[right_cave] == 1) or \
            (self.__puzzle == 2 and can_be_visited(right_cave, counter))):
                counter[right_cave] += 1
                self.__next_cave(right_cave, counter, path)
                counter[right_cave] -= 1
            elif right_cave == current and \
             (self.__puzzle == 1 and not (left_cave.lower() == left_cave and counter[left_cave] == 1) or
              (self.__puzzle == 2 and can_be_visited(left_cave, counter))):
                counter[left_cave] += 1
                self.__next_cave(left_cave, counter, path)
                counter[left_cave] -= 1
        return 0

    def traverse(self):
        local_counter = self.__cave_counter.copy()
        self.__paths = []
        self.__next_cave("start", local_counter, "")
        print(len(list(set(self.__paths))))
            
            
day12_test_1a = DAY_12("example_1.txt", puzzle=1, verbose=False)
day12_test_1a.read_map()
day12_test_1a.count_caves()
day12_test_1a.traverse()

day12_test_1b = DAY_12("example_2.txt", puzzle=1, verbose=False)
day12_test_1b.read_map()
day12_test_1b.count_caves()
day12_test_1b.traverse()

day12_test_1c = DAY_12("example_3.txt", puzzle=1, verbose=False)
day12_test_1c.read_map()
day12_test_1c.count_caves()
day12_test_1c.traverse()

day12_puzzle_1 = DAY_12("input.txt", puzzle=1, verbose=False)
day12_puzzle_1.read_map()
day12_puzzle_1.count_caves()
day12_puzzle_1.traverse()
```
