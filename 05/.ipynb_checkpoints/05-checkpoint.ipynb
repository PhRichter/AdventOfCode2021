```python
import numpy as np

class DAY_05:
    def __init__(self, log, size):
        self.__log = log
        self.__map = np.zeros((size, size))
        self.__x1 = []
        self.__x2 = []
        self.__y1 = []
        self.__y2 = []

    def read_vents(self):
        with open(self.__log, "r") as f:
            for line in f.readlines():
                self.__x1.append(line.split(",")[0])
                self.__y2.append(line.split(",")[-1])
                self.__y1.append(line.split(",")[1].split("->")[0])
                self.__x2.append(line.split(",")[1].split("->")[1])
        self.__x1 = np.array(self.__x1, dtype=int)
        self.__x2 = np.array(self.__x2, dtype=int)
        self.__y1 = np.array(self.__y1, dtype=int)
        self.__y2 = np.array(self.__y2, dtype=int)
        
    def fill_map_horizontal_vertikal(self):
        for i in range(self.__x1.size):
            #Horizontal
            if self.__x1[i] == self.__x2[i]:
                limits = np.sort([self.__y1[i], self.__y2[i]])
                for pos in np.arange(limits[0], limits[1]+1, 1):
                    self.__map[self.__x2[i], pos] += 1
                
            #Vertikal
            if self.__y1[i] == self.__y2[i]:
                limits = np.sort([self.__x1[i], self.__x2[i]])
                for pos in np.arange(limits[0], limits[1]+1, 1):
                    self.__map[pos, self.__y2[i]] += 1
                #return
                
    def fill_map_diagonal(self):
        for i in range(self.__x1.size):
            if self.__x1[i] - self.__y1[i] == self.__x2[i] - self.__y2[i]:
                vents = self.__x1[i] - self.__x2[i]
                if vents > 0:
                    for j in range(vents+1):
                        self.__map[self.__x2[i]+j, self.__y2[i]+j] += 1
                else:
                    for j in range(np.abs(vents)+1):
                        self.__map[self.__x1[i]+j, self.__y1[i]+j] += 1
                        
            elif self.__x1[i] + self.__y1[i] == self.__x2[i] + self.__y2[i]:
                vents = self.__x1[i] - self.__x2[i]
                if vents > 0:
                    for j in range(vents+1):
                        self.__map[self.__x1[i]-j, self.__y1[i]+j] += 1
                else:
                    for j in range(np.abs(vents)+1):
                        self.__map[self.__x2[i]-j, self.__y2[i]+j] += 1
            
            
    def find_dangerous_locations(self):
        print(np.where(self.__map >= 2)[0].size)

    def show_log(self):
        print("x1 ", self.__x1)
        print("x2 ", self.__x2)
        print("y1 ", self.__y1)
        print("y2 ", self.__y2)
        
    def show_map(self):
        print(self.__map.T)
        
day05_test_1 = DAY_05("example.txt", 10)
day05_test_1.read_vents()
day05_test_1.fill_map_horizontal_vertikal()
day05_test_1.show_map()
day05_test_1.find_dangerous_locations()

day05_puzzle_1 = DAY_05("input.txt", 1000)
day05_puzzle_1.read_vents()
day05_puzzle_1.fill_map_horizontal_vertikal()
day05_puzzle_1.find_dangerous_locations()
```

```python
day05_test_2 = DAY_05("example.txt", 10)
day05_test_2.read_vents()
day05_test_2.fill_map_horizontal_vertikal()
day05_test_2.fill_map_diagonal()
day05_test_2.find_dangerous_locations()
day05_test_2.show_map()

day05_puzzle_2 = DAY_05("input.txt", 1000)
day05_puzzle_2.read_vents()
day05_puzzle_2.fill_map_horizontal_vertikal()
day05_puzzle_2.fill_map_diagonal()
day05_puzzle_2.find_dangerous_locations()
```
